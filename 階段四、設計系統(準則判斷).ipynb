> ç¸½å ±é…¬æœ¬ç›Šæ¯”(final_eps) = ç¸½å ±é…¬ç‡(total_ROI) / æœ¬ç›Šæ¯”(pe_ratio_result)

> ç¸½å ±é…¬ç‡(total_ROI) = ç›ˆé¤˜æˆé•·ç‡(eps_yield) + æ”¶ç›Šç‡(div_yield)

> æ”¶ç›Šç‡(div_yield) = å…¨å¹´è‚¡æ¯(stkdiv) / è‚¡åƒ¹(lastprice) *100%
  
ä»¥ä¸Šè³‡æ–™æˆ‘å€‘åœ¨`éšæ®µä¸‰ã€è¨­è¨ˆç³»çµ±(æŸ¥è©¢å€‹è‚¡)`éƒ½å·²ç¶“å–å¾—éäº†ï¼Œä½†æ˜¯è¦å°å¿ƒå–å›æ­£ç¢ºæ­£å¸¸è³‡æ–™ï¼Œä¸ç„¶æœƒå½±éŸ¿åˆ°è¨ˆç®—çµæœã€‚

```python
divi_check = 1
eps_check = 1
stkdiv = 8
eps_yield = -64.85
characters = "%," #è¨­å®šç‰¹å®šå­—ç¬¦
pe_ratio_result = 19.62
if divi_check == 1: #diviæª¢æŸ¥æˆåŠŸ
    if eps_check == 1: #epsæª¢æŸ¥æˆåŠŸ
        div_yield = ("{:.2%}".format(stkdiv/lastprice)) #æ”¶ç›Šç‡ = (å…¨å¹´è‚¡æ¯/è‚¡åƒ¹*100%)
        for e in range(len(characters)): #å»é™¤æœƒå½±éŸ¿æ ¼å¼è½‰æ›çš„ç¬¦è™Ÿ
            div_yield = div_yield.replace(characters[e],"")
        div_yield = float(div_yield) #æ ¼å¼è½‰æ›æˆæµ®é»
        total_ROI = eps_yield + div_yield # (ç›ˆé¤˜æˆé•·ç‡+æ”¶ç›Šç‡ = ç¸½å ±é…¬ç‡)
        final_eps = (total_ROI/pe_ratio_result)/100 # ( ç¸½å ±é…¬ç‡/æœ¬ç›Šæ¯” = ç¸½å ±é…¬æœ¬ç›Šæ¯”)
print(final_eps)
```
